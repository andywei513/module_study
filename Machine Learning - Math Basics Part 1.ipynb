### 梯度及梯度下降法
`机器学习中大部分问题都是优化问题，而绝大部分优化问题都可以使用梯度下降法处理，因此掌握梯度下降法尤为重要。`

#### 1. 导数、偏导数和方向导数

- ***导数***：反映的是函数y=f(x)在某一点处沿x轴正方向的变化率。如果f’(x)>0，说明f(x)的函数值在x点沿x轴正方向是趋于增加的；如果f’(x)<0，说明f(x)的函数值在x点沿x轴正方向是趋于减少的。

<br/>
- ***偏导数***：反映的是函数y=f(x1,x2,…,xn)在某一点处沿某一坐标轴(x1,x2,…,xn)正方向的变化率。

<br/>
- ***方向导数***：反映的是函数y=f(x1,x2,…,xn)沿任意方向（含非坐标轴方向）的变化率，即某一点在某一趋近方向上的导数值。

<br/>

#### 2. 梯度（∇f）
&emsp;`梯度的提出是为了回答函数在变量空间的某一点处，沿着哪个方向有最大的变化率的问题。`<br/>
&emsp;**定义**：函数在某一点的梯度是这样一个向量，它的方向与取得最大方向导数的方向一致，而它的模为方向导数的最大值。向量的方向为变化率的方向，向量的模为变化率的大小。梯度即函数在某一点最大的方向导数，函数沿梯度方向函数有最大的变化率。



- 在单变量函数中，梯度为函数的微分，代表函数在某个给定点的切线的斜率。
- 在多变量函数中，梯度是一个向量，向量有方向，梯度的方向为函数在给定点的上升最快的方向。

&emsp;**特点**：
- 梯度是一个向量，即有方向和大小
- 梯度的方向是最大方向导数的方向
- 梯度的值是最大方向导数的值
